## STRONG SCALING (MPI)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Leggi il file CSV utilizzando Pandas
data = pd.read_csv("csvs\mpi_strong_thin.csv")

# Estrai le colonne
x = data["Total Tasks"]
y = data["Execution Time (s)"]

# Plotta i dati
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-')
plt.xlabel('Total Tasks')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time vs Total Tasks')

# Imposta l'asse y da 2.4 a 3.1 con passo 0.05
#plt.ylim(2.4, 3.1)
#plt.yticks(np.arange(2.4, 3.1, 0.05))

# Imposta l'asse x con incrementi di dieci
#plt.xticks(np.arange(min(x), max(x)+1, 10))

plt.grid(True)
plt.show()
```

## WEAK SCALING (MPI)

## WEAK SCALING (OMP)

```python
# Leggi il file CSV utilizzando Pandas
data = pd.read_csv("csvs/omp_weak_thin.csv")

# Estrai le colonne
threads = data["OMP_NUM_THREADS"]
execution_times = data["Execution_Time"]

# Plotta i dati
plt.figure(figsize=(10, 6))
plt.plot(threads, execution_times, marker='o', linestyle='-')
plt.xlabel('OMP_NUM_THREADS')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time vs OMP_NUM_THREADS')
plt.grid(True)
plt.show()
```
