BROADCAST 0

```python
import pandas as pd

# Read the CSV file into a pandas bcast0Frame
bcast0 = pd.read_csv("C:/Users/minut/OneDrive/Desktop/HPC_ex/HPC_Ex/output_processes_size_0_new.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Creare una palette di colori
palette = plt.get_cmap('tab10')

# Plot per ogni gruppo di 'Processes'
for i, process in enumerate(bcast0['Processes'].unique()):
    subset = bcast0[bcast0['Processes'] == process]
    plt.plot(subset['Size'], subset['Latency'], marker='', color=palette(i), linewidth=2, label=process)

# Aggiungere legenda, titolo e etichette agli assi
plt.legend(title='Processes')
plt.title('Latency vs Size per Different Processes')
plt.xlabel('Size')
plt.ylabel('Latency')

# Mostrare il grafico
#plt.xscale('log', base=2)  # Scala logaritmica per l'asse x per gestire meglio le dimensioni
plt.show()
```
