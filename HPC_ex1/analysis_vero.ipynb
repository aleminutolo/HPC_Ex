# BROADCAST NODE

## BCAST 1

```python
import pandas as pd

# Sostituisci 'path_to_file.csv' con il percorso effettivo del tuo file CSV
data = pd.read_csv('new_csvs/bcast1_fixed_node.csv')
print(data.head())
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Seleziona solo le colonne rilevanti
X = data[['Processes']]
y = data['Latency']

# Fit del modello di regressione
model = LinearRegression()
model.fit(X, y)

# Stampa i coefficienti
print(f'Intercept: {model.intercept_}')
print(f'Coefficient for Processes: {model.coef_[0]}')
```

```python
# Genera predizioni per la linea di regressione
x_range = np.linspace(X.min(), X.max(), 100)
y_preds = model.predict(x_range.reshape(-1, 1))

# Plot dei dati originali e della linea di regressione
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Dati Osservati')
plt.plot(x_range, y_preds, color='red', label='Linea di Regressione')
plt.title('Latenza vs Numero di Processi')
plt.xlabel('Numero di Processi')
plt.ylabel('Latenza (ms)')
plt.legend()
plt.show()
```
